````python
# python 模块化引入
# import OpenAi from openai
# js抄袭python
# gemini claude 接口不一样
from openai import OpenAI

# 替换为你的 DeepSeek API Key
client = OpenAI(
    api_key="api key",
    base_url="https://api.deepseek.com"
)
````

````python
# 参数默认值，es6借鉴了，函数灵活
# llm 很强大，封装的很好，就是几个接口
# completion 接口 aigc prompt
# chat 聊天接口 messages 多轮对话
# chat 分角色 user，assistant
def get_completion(prompt, model='deepseek-chat'):
    response = client.chat.completions.create(
        model=model,
        messages = [
            {"role": "user", "content": prompt}
        ],  
        # 自由度参数 0严谨 2嘻哈随意
        temperature=0.5,  # 修复：添加英文逗号
        # 问题+回复的tokens数量
        max_tokens=1024,
        n = 1
    )
    print(response)
    return response.choices[0].message.content

print(get_completion("写一首夏天游玩的七言诗歌"))
````

````python
text = f"""
你应当通过尽可能清晰、具体的指令，来明确你希望模型完成的任务。
这能引导模型产出符合预期的结果，同时降低回复内容偏离主题或出现错误的概率。
不要把编写清晰的提示词和精简提示词混为一谈。
很多时候，篇幅更长的提示词能为模型提供更完整的说明与背景信息，进而让输出内容更加详实、贴合需求。
"""
prompt = f"""
将三个反引号之间的文本总结为一句话
```{text}```
"""
response = get_completion(prompt)
print(response)
# 编写提示词应清晰具体，无需刻意精简，长提示词能提供更多信息使输出更准确贴合需求。
````

````python
prompt = f"""
请列出四大名著，并标注对应的作者与书籍类型
使用json格式呈现，需包含以下字段：book_id(书籍编号)，title(书名)，author(作者),desc(简介)genre(书籍类型)
"""
response = get_completion(prompt)
print(response)
````
